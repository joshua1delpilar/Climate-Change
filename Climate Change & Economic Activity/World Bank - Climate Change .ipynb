### Reorganising Data Year Columns


```python
def reorgYear(data):
    columnNames = ['Country code', 'Country name', 
                   'Series code', 'Series name', 'SCALE', 
                   'Decimals',]
    valueName = np.array(data.iloc[1:2, 3:4]).tolist()[0][0]
    newData = data.melt(id_vars=columnNames, var_name = 'Year', value_name=valueName)
    return newData
```

```python
# Completely Segragated Data
def completeSegData(data, mask):
    inputData = Data(data, mask)
    cleanedData = missingValues(inputData)
    segData = reorgYear(cleanedData)
    return segData
```

### Data Map 

```python
def mapGraph(year, inputData, mask):
    from IPython.display import HTML
    import folium
    # Data preparation
    inputData = inputData[mask]
    column_name1 = np.array(inputData.iloc[1:2, 3:4]).tolist()[0][0] 
    inputData = inputData.melt(id_vars = columnNames, var_name = 'Year', value_name = column_name1)  # Abundant Data
    maskYear = inputData['Year'] == year
    inputData = inputData[maskYear]
    inputData = inputData.replace(['..'], 0)
    column_name2 = inputData.columns[7]
    
    country_geo = 'world-countries.json'

    # Setup a folium map at a high level zoom
    map = folium.Map(location=[1,0], zoom_start = 1.5)

    # Geographic Emissions
    map.choropleth(geo_data=country_geo, data=inputData,
              columns = ['Country code', column_name2],
              key_on = 'feature.id',
              fill_color='BuPu',
              fill_opacity=0.7,
              line_opacity=0.5,
              legend_name= column_name2 + ' ' + str(year))
              
    file = map.save('co2_data.html')
    html = HTML('<iframe src=co2_data.html width=900 height=450></iframe>')
    return html
```
